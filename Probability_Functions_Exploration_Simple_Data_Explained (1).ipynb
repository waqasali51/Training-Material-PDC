## Probability Mass Function (PMF)

The Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value.

### Mathematical Representation

Mathematical Representation: If
ùëã
X is a discrete random variable, the PMF is

P(X=x).

### Example with a Simple Dataset

Consider a dataset representing the number of heads obtained when flipping a coin three times.

### Calculation and Graph

```python
# Simple dataset: number of heads in 3 coin flips
data = [0, 1, 2, 3]
counts = [1, 3, 3, 1]  # Corresponding counts from 3 coin flips

# Calculate PMF
total_flips = sum(counts)
pmf = [count / total_flips for count in counts]

plt.figure(figsize=(10, 6))
plt.stem(data, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Number of Heads in 3 Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```

In this example, the PMF shows the probability of getting 0, 1, 2, or 3 heads when flipping a coin three times.


````python

# Simple dataset: number of heads in 3 coin flips
data = [0, 1, 2, 3]
counts = [1, 3, 3, 1]  # Corresponding counts from 3 coin flips

# Calculate PMF
total_flips = sum(counts)
pmf = [count / total_flips for count in counts]

plt.figure(figsize=(10, 6))
plt.stem(data, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Number of Heads in 3 Coin Flips')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
````

### Example with a Simple Dataset

Consider a dataset representing the scores of students in a test.

### Calculation and Graph

```python
# Simple dataset: student scores
scores = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]

# Calculate CDF
sorted_scores = np.sort(scores)
cdf = np.arange(1, len(sorted_scores) + 1) / len(sorted_scores)

plt.title('CDF of Student Scores')
plt.xlabel('Score')
plt.ylabel('Cumulative Probability')
plt.show()
```

In this example, the CDF shows the cumulative probability of a student scoring less than or equal to a particular score.


````python

# Simple dataset: student scores
scores = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100]

# Calculate CDF
sorted_scores = np.sort(scores)
cdf = np.arange(1, len(sorted_scores) + 1) / len(sorted_scores)

plt.figure(figsize=(10, 6))
plt.step(sorted_scores, cdf, where='post')
plt.title('CDF of Student Scores')
plt.xlabel('Score')
plt.ylabel('Cumulative Probability')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simple dataset: heights of individuals (in cm)
heights = [150, 155, 160, 165, 170, 175, 180, 185, 190, 195]

# Step 1: Calculate Mean and Standard Deviation
mean = np.mean(heights)
std_dev = np.std(heights)

# Step 2: Define the range of values for which we want to estimate the PDF
x_values = np.linspace(min(heights) - 10, max(heights) + 10, 1000)

# Step 3: Calculate the PDF using the normal distribution formula
pdf_values = norm.pdf(x_values, mean, std_dev)

# Plotting the PDF
plt.plot(x_values, pdf_values, color='blue', label='PDF')
plt.xlabel('Height (cm)')
plt.ylabel('Density')
plt.title('Estimated PDF of Heights')
plt.legend()
plt.grid(True)
plt.show()
````

## Combined Example

### Example with a Simple Dataset

Consider a dataset representing the ages of individuals in a small group.

### Calculation and Graph

```python
# Simple dataset: ages of individuals
ages = [23, 23, 24, 24, 24, 25, 25, 26, 26, 27, 28, 29, 30]

# PMF
values, counts = np.unique(ages, return_counts=True)
pmf = counts / len(ages)

plt.figure(figsize=(18, 6))

# PMF Plot
plt.subplot(1, 3, 1)
plt.stem(values, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Ages')
plt.xlabel('Age')
plt.ylabel('Probability')

# CDF
sorted_ages = np.sort(ages)
cdf = np.arange(1, len(sorted_ages) + 1) / len(sorted_ages)
plt.subplot(1, 3, 2)
plt.step(sorted_ages, cdf, where='post')
plt.title('CDF of Ages')
plt.xlabel('Age')
plt.ylabel('Cumulative Probability')

# PDF
kde = gaussian_kde(ages)
x = np.linspace(min(ages), max(ages), 100)
pdf = kde(x)
plt.subplot(1, 3, 3)

plt.ylabel('Density')

```

In this example, the PMF, CDF, and PDF of the ages dataset are calculated and visualized for a comprehensive understanding of these functions.


````python

# Simple dataset: ages of individuals
ages = [23, 23, 24, 24, 24, 25, 25, 26, 26, 27, 28, 29, 30]

# PMF
values, counts = np.unique(ages, return_counts=True)
pmf = counts / len(ages)

plt.figure(figsize=(18, 6))

# PMF Plot
plt.subplot(1, 3, 1)
plt.stem(values, pmf, basefmt=" ", use_line_collection=True)
plt.title('PMF of Ages')
plt.xlabel('Age')
plt.ylabel('Probability')

# CDF
sorted_ages = np.sort(ages)
cdf = np.arange(1, len(sorted_ages) + 1) / len(sorted_ages)
plt.subplot(1, 3, 2)
plt.step(sorted_ages, cdf, where='post')
plt.title('CDF of Ages')
plt.xlabel('Age')
plt.ylabel('Cumulative Probability')

# PDF
kde = gaussian_kde(ages)
x = np.linspace(min(ages), max(ages), 100)
pdf = kde(x)
plt.subplot(1, 3, 3)
plt.plot(x, pdf, 'blue')
plt.title('PDF of Ages')
plt.xlabel('Age')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
````
